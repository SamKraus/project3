# Create the Dataframe

```python
import pandas as pd
import numpy as mp
```

```python
# Create a reference to the CSV and import it into a Pandas DataFrame
csv_path = "Resources/student-por.csv"
students_df = pd.read_csv(csv_path)
students_df.head()
```

```python
#students_df = students_f[students_f.romantic != 'yes']
#students_df
```

```python
students_df.describe()
```

```python
students_df.columns
```

```python
students_df["overall_alc"] = (students_df["Dalc"]*5+students_df["Walc"]*2)/7
students_df["overall_alc"] = students_df["overall_alc"].apply(lambda x: round(x,0))
   


students_df.head()
```

```python
# Drop unrelated columns 
students_df = students_df.drop(['school','G1', 'G2', 'G3', 'Dalc', 'Walc'], axis=1)
students_df.head()
```

```python
students_df.dtypes
```

```python
students_df.head()
```

```python
# perform get dummies on data set to binary encode for scaling
encoded_students_df = pd.get_dummies(students_df)
encoded_students_df.head()
```

```python
# keep encoded df clean from y data for prediction purposes
working_students_df = encoded_students_df.copy()
working_students_df.head()
```

```python
overall_alc_col = working_students_df[['overall_alc']]
new_df = working_students_df.drop(['overall_alc'], axis=1)
new_df['overall_alc'] = overall_alc_col
new_df.head()
```

```python
new_df.columns
```

```python
# Create the bins in which Drinking levels will be held
bins = [0,1,2,3,4,5]

# Create the names for the five bins
group_names = ["a.Not at all", "b.Slightly", "c.Moderately", "d.Very", "e.Extremely"]
```

```python
# Create column to contain new bins
new_df["Likelyhood to Drink"] = pd.cut(new_df["overall_alc"], bins, labels=group_names)
new_df = new_df.drop(['overall_alc'], axis=1)

new_df
```

```python
# Reformat data (x-inputs, y labels)
data = new_df.values
X = data[:, 0:52]
y = data[:, 52]

y = y.reshape(-1, 1)
print(X)
print(y)
```

```python
print(X.shape)
print(y.shape)
```

# Data Processing

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)

print(encoded_y)
```

```python
for label, original_class in zip(encoded_y, y):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, encoded_y, random_state=1, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
X_scaler = StandardScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from keras.utils import to_categorical

# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
y_train_categorical
```

# Create Deep Learning Model

```python
from tensorflow.keras.models import Sequential

model = Sequential()
```

```python
from tensorflow.keras.layers import Dense
number_inputs = 52

number_hidden_nodes = 100
model.add(Dense(units=number_hidden_nodes,
                activation='relu', input_dim=number_inputs))
```

```python
# add another layer??

number_layer_classes = 100
model.add(Dense(units=number_layer_classes, activation='relu'))
```

```python
# add another layer??
number_classes = 5
model.add(Dense(units=number_classes, activation='softmax'))
```

```python
model.summary()
```

```python
# Use categorical crossentropy for categorical data and mean squared error for regression
# Hint: your output layer in this example is using software for logistic regression (categorical)
# If your output layer activation was `linear` then you may want to use `mse` for loss
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit (train) the model
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
# Evaluate the model using the testing data
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model.predict_classes(X_test_scaled[:5])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:5])}")
```

```python
#Save the model
model.save("students_deep_learning_dalcwalc_trained.h5")
```
