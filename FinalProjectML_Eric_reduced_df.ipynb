```python
# Drop unrelated columns and Walc
students_df = students_df.drop(['school','G1', 'G2', 'G3', 'Walc'], axis=1)
students_df.head()
```

```python
reduced_students_df = students_df.drop(['traveltime', 'studytime', 'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'romantic', 'famrel', 'freetime', 'goout', 'health', 'absences'], axis=1)
reduced_students_df.head()

# 'school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu',
#        'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime',
#        'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
#        'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'Dalc',
#        'Walc', 'health', 'absences', 'G1', 'G2', 'G3'
```

```python
# perform get dummies on data set (less Walc) to binary encode for scaling
encoded_reduced_students_df = pd.get_dummies(reduced_students_df)
encoded_reduced_students_df.head()
```

```python
# keep encoded df clean from y data for prediction purposes
working_reduced_students_df = encoded_reduced_students_df.copy()
working_reduced_students_df.head()
```

```python
Dalc_col = working_reduced_students_df[['Dalc']]
new_reduced_df = working_reduced_students_df.drop(['Dalc'], axis=1)
new_reduced_df['Dalc'] = Dalc_col
new_reduced_df.head()
```

```python
# Create the bins in which Drinking levels will be held
bins = [0,1,2,3,4,5]

# Create the names for the five bins
group_names = ["A.Not at all", "B.Slightly", "C.Moderately", "D.Very", "E.Extremely"]
```

```python
# # Create column to contain new bins
new_reduced_df["Likelyhood to Drink"] = pd.cut(new_reduced_df["Dalc"], bins, labels=group_names)
new_reduced_df = new_reduced_df.drop(['Dalc'], axis=1)

new_reduced_df
```

```python
# Reformat data (x-inputs, y labels)
data = new_reduced_df.values
X = data[:, 0:30]
y = data[:, 30]

y = y.reshape(-1, 1)
# print(X)
print(y)
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()

label_encoder.fit(y)
encoded_y = label_encoder.transform(y)

print(encoded_y)
```

```python
for label, original_class in zip(encoded_y, y):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,encoded_y, random_state=1)
```

## Data Preprocessing ##

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
X_scaler = StandardScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from keras.utils import to_categorical

# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

## Create Deep Learning Model ##

```python
from tensorflow.keras.models import Sequential

model = Sequential()
```

```python
from tensorflow.keras.layers import Dense
number_inputs = 30

number_hidden_nodes = 156
model.add(Dense(units=number_hidden_nodes,
                activation='relu', input_dim=number_inputs))
```

```python
# add another layer??
number_layer_classes = 5
model.add(Dense(units=number_layer_classes, activation='relu'))
model.add(Dense(units=number_layer_classes, activation='relu'))
# model.add(Dense(units=number_layer_classes, activation='relu'))
# model.add(Dense(units=number_layer_classes, activation='relu'))
# model.add(Dense(units=number_layer_classes, activation='relu'))
```

```python
# add another layer??
number_classes = 5
model.add(Dense(units=number_classes, activation='softmax'))
```

```python
model.summary()
```

```python
# Use categorical crossentropy for categorical data and mean squared error for regression
# Hint: your output layer in this example is using software for logistic regression (categorical)
# If your output layer activation was `linear` then you may want to use `mse` for loss
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
X_train_scaled.shape=(486,30)
```

```python
# Fit (train) the model
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=40,
    shuffle=True,
    verbose=2
)
```

## Quantify Trained Model ##

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
